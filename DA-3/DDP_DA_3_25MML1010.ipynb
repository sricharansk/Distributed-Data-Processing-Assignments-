🔹 What is Structured Streaming?

  * Structured Streaming in Apache Spark is a real-time data processing engine built on Spark SQL.
  * It treats streaming data as a continuous DataFrame/Table and processes it incrementally.

🔹 Difference: Structured Streaming vs Batch Processing

   | Feature    | Structured Streaming   | Batch Processing     |
   | ---------- | ---------------------- | -------------------- |
   | Data Type  | Continuous (real-time) | Static (stored data) |
   | Processing | Ongoing                | One-time             |
   | Latency    | Low (near real-time)   | High                 |
   | Example    | Live logs, sensors     | Historical reports   |

🔹 Key Features

  * Unified API (same as batch → DataFrame API)
  * Fault-tolerant (recovers from failures)
  * Scalable (handles large streams)
  * Exactly-once processing (no data duplication)
  * Event-time processing (handles late data)

🔹 Real-Time Use Cases

  * Live log monitoring (server logs)
  *  Fraud detection in banking
  * Social media analytics
  * IoT sensor data processing

```python
import os
```

```python
from pyspark.sql import SparkSession

# Step 1: Create Spark Session
spark = SparkSession.builder \
    .appName("StructuredStreamingColab") \
    .getOrCreate()

print("Spark Session Created")
```

```python
# Step 2: Create directory for streaming input files
input_path = "/content/stream_data"
os.makedirs(input_path, exist_ok=True)

print("Streaming folder created at:", input_path)
```

```python
# Step 3: Read streaming data from folder
# Spark continuously monitors this folder
lines = spark.readStream \
    .format("text") \
    .load(input_path)

print("Streaming source initialized")
```

```python
from pyspark.sql.functions import split, explode

# Step 4: Transform data
# Split each line into words and flatten
words = lines.select(
    explode(split(lines.value, " ")).alias("word")
)

# Count occurrences of each word
word_counts = words.groupBy("word").count()

print("Transformation applied (Word Count)")
```

```python
# Step 5: Output results to console
query = word_counts.writeStream \
    .outputMode("complete") \
    .format("console") \
    .start()

print("Streaming started...")
```

```python
# Step 6: Add files dynamically (simulate streaming)
import time

# First file
with open("/content/stream_data/data1.txt", "w") as f:
    f.write("hello spark hello world")

time.sleep(5)

# Second file
with open("/content/stream_data/data2.txt", "w") as f:
    f.write("spark streaming is powerful")

print("Streaming data added")
```

```python
# Step 7: Allow stream to process data
import time
time.sleep(10)

# Stop streaming
query.stop()
spark.stop()

print("Streaming stopped")
```
