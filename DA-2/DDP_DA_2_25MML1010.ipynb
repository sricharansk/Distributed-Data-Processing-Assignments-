Transformations vs Actions in Apache Spark (Brief Answer)

🔹 Transformations

  * Transformations are operations that create a new DataFrame/RDD from existing data.
  * They are lazy → not executed immediately.
  * Spark only builds an execution plan (DAG).

✔ Examples:

  * select() → selects specific columns
  * filter() → filters rows based on condition
  * groupBy() → groups data for aggregation

🔹 Significance in Execution Planning

  * Spark builds a DAG (Directed Acyclic Graph) of all transformations
  * Optimizes the plan (removes unnecessary steps, combines operations)
  * Improves performance before actual execution
🔹 Actions
  * Actions are operations that trigger execution of the DAG.
  * They produce final output or result.

✔ Examples:

  * show() → displays data
  * count() → returns number of rows
  * collect() → brings data to driver

🔹 Significance in Execution Planning

  * Actions start the execution process
  * Spark executes all transformations in an optimized way
  * Produces actual results after computation

⚡ Key Difference (Quick View)

| Feature   | Transformations      | Actions                |
| --------- | -------------------- | ---------------------- |
| Execution | Lazy                 | Immediate              |
| Purpose   | Build execution plan | Execute plan           |
| Output    | New DataFrame/RDD    | Result (value/display) |



```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Assignment2").getOrCreate()

# Sample dataset
data = [
    (1, "Alice", "HR", 30000),
    (2, "Bob", "IT", 40000),
    (3, "Charlie", "IT", 35000),
    (4, "David", "HR", 28000),
    (5, "Eva", "Finance", 50000)
]

columns = ["ID", "Name", "Department", "Salary"]

df = spark.createDataFrame(data, columns)
```

```python
df_select = df.select("Name", "Department")
```

```python
df_filter = df.filter(df.Salary > 30000)
```

```python
df_group = df.groupBy("Department").count()
```

```python
print("Selected Columns:")
df_select.show()
```

```python
print("Filtered Data (Salary > 30000):")
df_filter.show()
```

```python
print("Group By Department:")
df_group.show()
```
