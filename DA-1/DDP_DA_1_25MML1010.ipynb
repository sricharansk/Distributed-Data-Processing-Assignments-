**What is Distributed Data Processing?**


*   Distributed processing means splitting large data into smaller chunks and processing them across multiple machines (nodes) in parallel
*   Instead of one system doing all work, a cluster of computers works together, improving speed and scalability.
*   Common in Big Data systems where data is too large for a single machine.

**How Apache Spark Improves Performance?**

  * Apache Spark is a fast, distributed data processing framework.

  * Key improvements over traditional systems (like disk-based processing engines):

     * In-Memory Computation: Stores data in RAM → avoids repeated disk I/O → much faster.

     * Lazy Evaluation: Executes operations only when needed → optimizes execution plan.

     * Parallel Processing: Runs tasks simultaneously across nodes.    

     * Fault Tolerance: Uses lineage (DAG) to recompute lost data.

**Role of In-Memory Computation**

 *  Traditional systems store intermediate results on disk (slow)

 *  Spark stores data in memory (RAM)

Result:

 *  Faster processing (10x–100x speed improvement)

 *  Efficient for iterative tasks (ML, analytics)

Write a PySpark program to:

- Create a DataFrame with at least 5 records (columns: ID, Name, Age, Salary)  

- Display the schema

- Show all records

- Filter employees with salary greater than 30,000

```python
# Import SparkSession
from pyspark.sql import SparkSession
```

```python
# Create Spark session
spark = SparkSession.builder.appName("EmployeeData").getOrCreate()
```

```python
# Sample data (ID, Name, Age, Salary)
data = [
    (1, "Alice", 25, 35000),
    (2, "Bob", 30, 28000),
    (3, "Charlie", 35, 40000),
    (4, "David", 28, 25000),
    (5, "Eva", 32, 50000)
]
```

```python
# Define column names
columns = ["ID", "Name", "Age", "Salary"]
```

```python
# Create DataFrame
df = spark.createDataFrame(data, columns)
```

```python
# Display schema
print("Schema:")
df.printSchema()
```

```python
# Show all records
print("All Records:")
df.show()
```

```python
# Filter employees with salary > 30000
print("Employees with Salary > 30000:")
df.filter(df.Salary > 30000).show()
```

```python
# Stop Spark session
spark.stop()
```
